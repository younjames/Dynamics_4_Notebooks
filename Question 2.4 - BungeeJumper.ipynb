Obtain the equations of motion of the system considering it is released from the horizontal position at time $t=0$.

## <span style="color:#0073C0">Example Solution</span>
Let us start by considering thatais the initial length of the spring (bungee cord) and that at the start $\theta=0$ and $x=0$. After time $t$ the mass has a radial velocity ̇$x$ which is the velocity of $m$ relative to the bar. The tangential component of the velocity is $(a+x)\dot{\theta}$. Consequently, the kinetic energy of the system is
$$ T = \frac{1}{2}m\left[\dot{x}^2+(a+x)^2\dot{\theta}^2\right] $$
The potential energy of the system is
$$V = \frac{1}{2}kx^2-mg(a+x)sin\theta$$
As there is no drag, there is no generalised force on the system and Lagrange’s equations are

$$\frac{\mathrm{d}}{\mathrm{d}t}\left(\frac{\partial T}{\partial \dot{x}}\right)-\frac{\partial T}{\partial x} + \frac{\partial V}{\partial x} = 0$$

$$\frac{\mathrm{d}}{\mathrm{d}t}\left(\frac{\partial T}{\partial \dot{\theta}}\right)-\frac{\partial T}{\partial \theta} + \frac{\partial \theta}{\partial x} = 0$$

Differentiating the energy expressions leads to

$$\frac{\partial T}{\partial \dot{\theta}}=m(a+x)^2\dot{\theta}\; \quad \frac{\mathrm{d}}{\mathrm{d}t}\left(\frac{\partial T}{\partial \dot{\theta}}\right) = m(a+x)^2\ddot{\theta} + 2m(a+x)^2\dot{x}\dot{\theta}; \quad \frac{\partial T}{\partial \theta}=0$$

$$\frac{\partial T}{\partial \dot{x}}=m\dot{x}\; \quad \frac{\mathrm{d}}{\mathrm{d}t}\left(\frac{\partial T}{\partial \dot{x}}\right) = m\ddot{x}\; \quad \frac{\partial T}{\partial x}=m(a+x)\dot{\theta}^2$$

$$ \frac{\partial V}{\partial \theta}=-mg(a+x)\cos\theta $$

$$ \frac{\partial V}{\partial x}=kx - mg\sin\theta $$

Consequently, the equations of motion for this system are:

$$m(a+x)^2\ddot{\theta} + 2m(a+x)\dot{x}\dot{\theta}-mg(a+x)\cos\theta=0$$
$$m\ddot{x}-m(a+x)\dot{\theta}^2+kx-mg\sin{\theta}=0$$

## <span style="color:#0073C0">Solving the equations of motion using Python (step-by-step)</span>
To solve these equations we start by importing the necessary python libraries.

```python
from pylab import *
from scipy.integrate import odeint
```

### Variables, constants and initial conditions
To solve this, we need to track four parameters, the linear and angular positions and velocities of mass $m$. To do this we create an array `y` with four elements, to store these four parameters. Now we define the initial conditions, which can be changed to study the behaviour of the system in different conditions:
 - $N$ (`N`) is the total number of steps, change this to increase/decrease resolution of your plot;
 - $a$ (`aa`) [m] is the unstreched length of the bungee cable;
 - $x_0$ (`x_0`) [m] is the initial strech of the cable;
 - $v_0$ (`v_0`) [m/s] is the initial (release) velocity;
 - $\theta_0$ (`theta_0`) [rad] is the initial (release) angle; and
 - $\omega_0$ (`omega_0`) [rad] is the initial (release) angular velocity.

```python
N = 1000
y = zeros([4])

aa = 8.0  
x_0 = 1.0
v_0 = 0.0
theta_0 = 0.5
omega_0 = 0.0

y[0] = x_0
y[1] = v_0
y[2] = theta_0
y[3] = omega_0
```

It is also necessary to define the remaining variables and constants in the equations of motion:
 - $t$ (`time`) is the time space for the calculations [s];
 - $k$ (`k`) is the cable stiffness (spring constant) [N/m];
 - $m$ (`m`) is the mass of the jumper [kg]; and
 - $g$ (`gravity`) is the acceleration of gravity [m/s$^2$].

```python
time = linspace(0, 25, N)
k = 150.0
m = 70.0
gravity = 9.8
```

### Solving the equations of motion

The equations of motion of the system can be algebraically manipulated and rewritten as

$$ \ddot\theta = \frac{\mathrm{d}^2\theta}{\mathrm{d}t^2} = \frac{1}{(a+x)}\left[g\cos\theta - 2\left(\frac{\mathrm{d}x}{\mathrm{d}t}\right)\left(\frac{\mathrm{d}\theta}{\mathrm{d}t}\right)\right] $$

$$ \ddot x = \frac{\mathrm{d}^2x}{\mathrm{d}t^2}  = (a+x)\left(\frac{\mathrm{d}\theta}{\mathrm{d}t}\right)^2 - \frac{k}{m}x + g\sin\theta $$

The `spring_pendulum` function defines this set of differential equations:

```python
def spring_pendulum(y, time):
    g0 = y[1]
    g1 = (aa + y[0])*y[3]*y[3] - k/m*y[0] + gravity*sin(y[2])
    g2 = y[3]
    g3 = (gravity*cos(y[2]) - 2.0*y[1]*y[3])/(aa + y[0])
    
    return array([g0, g1, g2, g3])
```

Now we find the solution to the equations of motion using the `odeint` python module from `scipy`. This module can be used to integrate a system of ordinary differential equations. You can find more information about this module in this <a href="https://docs.scipy.org/doc/scipy/reference/generated/scipy.integrate.odeint.html">Link</a>.

```python
answer = odeint(spring_pendulum, y, time)
```

### Plotting the results

The final step is to plot the results. Rather than graph in terms of $t$, we will track the position of the jumper in 2D. This will require that we change $(a+x,\theta)$ data to the two-dimensional $(x,y)$ space. The red dot marks the initial (release) position.

```python
xdata = (aa + answer[:,0])*cos(answer[:,2])
ydata = -(aa + answer[:,0])*sin(answer[:,2])
xIn = (aa+x_0)*cos(theta_0)
yIn = -(aa+x_0)*sin(theta_0)
plot(xdata,ydata, color='#0073C0', linewidth=1.)
plot(xIn, yIn, marker='o', markersize=8, markerfacecolor='red', markeredgewidth=0.7, markeredgecolor='black')
grid(color='grey', linestyle=':', linewidth=0.3)
xlabel('x [m]')
ylabel('y [m]')
```

## <span style="color:#0073C0">Interactive animation</span>
Scroll to the bottom of this page to find an interactive plot where you can vary the input variables and see the resulting path taken by the bungee jumper. This code still solves the equations of motion which you derived for the system given in the tutorial question:

$$m(a+x)^2\ddot{\theta} + 2m(a+x)\dot{x}\dot{\theta}-mg(a+x)\cos\theta=0$$
$$m\ddot{x}-m(a+x)\dot{\theta}^2+kx-mg\sin{\theta}=0$$

You don't need to inspect the code other than for interest.

```python
%matplotlib notebook
from scipy.integrate import odeint
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.animation as animation

"""
The pendulum equations of motion to be solved by the "odeint" function
"""
def spring_pendulum(y, time, aa, k, m, gravity):
    g0 = y[1]
    g1 = (aa + y[0])*y[3]*y[3] - k/m*y[0] + gravity*np.sin(y[2])
    g2 = y[3]
    g3 = (gravity*np.cos(y[2]) - 2.0*y[1]*y[3])/(aa + y[0])
    
    return np.array([g0, g1, g2, g3])

"""
A wrapper fuction which feeds in the variables, solves the ODEs, and returns the postional data
"""
def get_response(y, aa, k, m, gravity):
    x_0 = y[0]
    theta_0 = y[2]
    answer = np.empty(len(time))
    answer = odeint(spring_pendulum, y, time, args=(aa, k, m, gravity))
    xdata = (aa + answer[:,0])*np.cos(answer[:,2])
    ydata = -(aa + answer[:,0])*np.sin(answer[:,2])
    xIn = (aa+x_0)*np.cos(theta_0)
    yIn = -(aa+x_0)*np.sin(theta_0)
    
    return xdata, ydata, xIn, yIn

"""
A function to update the position of the cable and the circle on 
the animation (called every time that the sliders are changed) 
"""
def update_position(x_0, aa, theta_0):
    rmax = aa+x_0
    circ_size = 0.04*(aa+x_0)
    xpos = (rmax)*np.cos(-theta_0)
    ypos = (rmax)*np.sin(-theta_0)
    cable.set_xdata((0,xpos))
    cable.set_ydata((0,ypos))    
    circle.radius = circ_size
    circle.center = xpos, ypos

"""
The function that is called every time the sliders are changed. This stores the initial values,
calls get_response to solve the ode and return the positional data. Also upates the initial position
on the animation and deletes any existing trace lines.
"""
def update(x_0 = 1, v_0 = 0, theta_0 = 0.5, omega_0 = 0, aa=8.0, k=150.0, m=70.0, gravity=9.8):
    global xdata, ydata
    y[0] = x_0
    y[1] = v_0
    y[2] = theta_0
    y[3] = omega_0
    xdata, ydata, xIn, yIn = get_response(y, aa, k, m, gravity)
    update_position(x_0, aa, theta_0)
    fig.canvas.draw_idle()
    del ax_an.lines[1:]
    back_line, = ax_an.plot(xdata, ydata, 'grey', linewidth=1., alpha=0.4)
    ax_an.relim()
    ax_an.autoscale_view()
    ani.frame_seq = ani.new_frame_seq() 

"""
The initialisation function called by animation.FuncAnimation. Not strictly needed as blit has been
set to false
"""
def init_fn():
    cable.set_data([], [])
    line2.set_data([], [])
    circle.radius = circ_size
    circle.center = xpos, ypos
    return cable, line2, circle
   
"""
The function called by animation.FuncAnimation.This sets the cable and circle position and traces
the circle's positional history.
"""
def animate(i):
    cable.set_xdata((0,xdata[i]))
    cable.set_ydata((0,ydata[i]))
    circ_size_local = 0.04*(aa+x_0)
    circle.center = xdata[i], ydata[i]
    line2, = ax_an.plot(xdata[i-1:i+1], ydata[i-1:i+1], color='#0073C0', linewidth=0.5)
    return cable, line2, circle

# Setting up initial value for plots 
N = 1000
y = np.zeros([4])
aa = 8.0  
x_0 = 1.0
v_0 = 0.0
theta_0 = 0.5
omega_0 = 0.0
y[0] = x_0
y[1] = v_0
y[2] = theta_0
y[3] = omega_0
time = np.linspace(0, 25, N)
k = 150.0
m = 70.0
gravity = 9.8
y_init = y

# Calculating the initial response
xdata, ydata, xIn, yIn = get_response(y, aa, k, m, gravity)

#Setting up the figure
fig = plt.figure(figsize=(6,4))

#Setting up axes 2 with an initial plot (might not be needed)
ax_an = fig.add_subplot(1, 1, 1)
rmax = aa+x_0
xpos = (rmax)*np.cos(-theta_0)
ypos = (rmax)*np.sin(-theta_0)
cable, = ax_an.plot((0,xpos), (0,ypos), 'grey', linewidth=1., zorder=2)
back_line, = ax_an.plot(xdata, ydata, 'grey', linewidth=1., alpha=0.4, zorder=1)
ckwags = {"fill": True, "color": "grey"}
circ_size = 0.04*(aa+x_0)
circle = ax_an.add_patch(mpl.patches.Circle((xpos,ypos),circ_size,zorder=3,**ckwags))
plt.grid(color='grey', linestyle=':', linewidth=0.3)
# ax_an.set_aspect('equal') #Uncomment this line to make the aspect equal (figure changes size)
ax_an.relim()
ax_an.autoscale_view()
line2, = ax_an.plot(xdata[0], ydata[0], color='#0073C0', linewidth=0.5)
xlabel('x [m]')
ylabel('y [m]')
title('Position of the bungee jumper')
plt.tight_layout()

# Setting up the animation function, giving it the figure handle and "animate" function as arguments
ani = animation.FuncAnimation(fig, animate, init_func=init_fn, interval=7, blit=True)

# Setting up all the slides with descriptions and ranges
x_0_slide = widgets.FloatSlider(value=1.0, description='x_0', min=0.1, max=10, step=0.1)
v_0_slide = widgets.FloatSlider(value=0.0, description='v_0', min=-5.0, max=5.0, step=0.1)
theta_0_slide = widgets.FloatSlider(value=0.5, description='theta_0', min=-0.5, max=0.5, step=0.1)
omega_0_slide = widgets.FloatSlider(value=0.0, description='omega_0', min=-0.5, max=0.5, step=0.1)
aa_slide = widgets.FloatSlider(value=8.0, description='a', min=0, max=20, step=0.1)
k_slide = widgets.FloatSlider(value=150, description='k', min=100, max=200, step=0.1)
m_slide = widgets.FloatSlider(value=70, description='m', min=10, max=100, step=0.1)
gravity_slide = widgets.FloatSlider(value=9.8, description='gravity', min=0.1, max=20, step=0.1)
ui_1 = widgets.HBox([aa_slide, theta_0_slide, k_slide])
ui_2 = widgets.HBox([x_0_slide, omega_0_slide, gravity_slide])
ui_3 = widgets.HBox([v_0_slide, m_slide])
ui = widgets.VBox([ui_1,ui_2, ui_3])

# Telling the widget what to do when the sliders are changed, in this case calling the "update" function
out = widgets.interactive_output(update, {'x_0': x_0_slide, 'v_0': v_0_slide, 'theta_0': theta_0_slide,
                                          'omega_0':omega_0_slide,'aa': aa_slide, 'k': k_slide,
                                          'm': m_slide, 'gravity':gravity_slide,})
display(ui, out)

```

As a reminder,
 - $a$ (`aa`) [m] is the unstreched length of the bungee cable;
 - $x_0$ (`x_0`) [m] is the initial strech of the cable;
 - $v_0$ (`v_0`) [m/s] is the initial (release) velocity;
 - $\theta_0$ (`theta_0`) [rad] is the initial (release) angle;
 - $\omega_0$ (`omega_0`) [rad] is the initial (release) angular velocity;
 - $m$ (`m`) [kg] is the mass of the jumper;
 - $k$ (`k`) [N/m] is the bungee cord spring stiffness; and
 - $g$ (`gravity`) [m/s^2] is the acceleration due to gravity.
